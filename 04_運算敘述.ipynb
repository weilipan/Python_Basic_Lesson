> ### 防禦性寫法
請注意 <code>==</code> 是 *比較* 運算子, 但 <code>=</code> 是 *賦值* 運算子.

為避免"="(賦值)與"=="(判斷)弄混，可以將常數寫在前面

<code>a==10</code> 可以寫成 <code>10==a</code>

<em>這樣寫的好處是一旦將"="(賦值)與"=="(判斷)弄混會出現錯誤訊息。</em>

# 4.連鎖比較運算子 Chained Comparison Operators

Python與C/C++有些差異，它可以直接使用連鎖比較運算子來處理較為複雜的表達式，與一般數學運算式較為相同而且直觀。

這個單元我們將學習到使用連鎖運算子，也會結合**and** and **or**的使用。

```python
1 < 2 < 3
```

上面的敘述會判斷兩件事，分別為**1是否小於2**，**2是否小於3**。我們也可以使用 **and** 來處理。

```python
1<2 and 2<3
```

```python
# 這樣寫可以執行，但有誤導的嫌疑...
1 < 3 > 2
```

```python
# 上面的敘述可以寫成下面這樣：
1<3 and 3>2
```

```python
# 使用or進行連鎖
1==2 or 2<3
```

```python
# **or** 連鎖的敘述中其中之一成立時，結果則為True:
1==1 or 100==1
```
